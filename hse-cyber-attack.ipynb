## Learnings
***

Recommendations that all organisations should consider in the light of the experience of the HSE, in order to learn lessons from this Incident more broadly. They are not intended to be exhaustive, but act as an instructive set of learnings to consider in response to this Incident.

### Governance and cybersecurity leadership
***

###### Understanding of technology dependency and governance of technology risk
<hr width=50%;>

- understand the extent to which critical operations are dependent on technology [0]

- resiliency to widespread technology failure or compromise from an attack (which may occur indirectly through the supply chain) [0]

###### Cybersecurity strategy and leadership
<hr width=50%;>

- having a cybersecurity strategy [0]

- an element of the strategy should be consideration of the cyber risk posed by legacy IT [0]

< [Table of Contents](#top) | [References](#references) >

### Effective cybersecurity capability
***

###### Ransomware-specific assessment
<hr width=50%;>

- cybersecurity controls that should be assessed include:

i. sufficiency of security monitoring to detect and contain ransomware attacks in the early stages


ii. ability to prevent and detect the compromise of ``privileged`` (e.g. systems administrator) accounts


iii. the robustness of user authentication [0]

- publicly available frameworks and guidance are available from organisations such as the ``Cybersecurity & Infrastructure Security Agency`` (``CISA``) in the USA [0]

###### Effective cybersecurity monitoring and response
<hr width=50%;>

- detect and respond to the tools and techniques used by human-operated ransomware groups [0]

- deploying a capable ``Endpoint Detection & Response`` tool to detect and prevent malicious activity on workstations (fed by current cyber threat intelligence) [0]

- ensuring the development of skilled resources and processes so that security alerts are rapidly triaged, investigated and responded to [0]

###### Testing of cybersecurity capability through simulated attacks
<hr width=50%;>

- the use of ethical hackers simulating end-to-end attack techniques (i.e. ‘red team’ testing) - provides a threat-based perspective of an organisation’s vulnerability to ransomware and other types of attacks [0]

- identify and prioritise key security improvement areas and ensure that the organisation can effectively detect common attacker tools [0]

### Preparedness to respond and recover
***

###### Cybersecurity-specific incident response and crisis management plans
<hr width=50%;>

- develop and exercise cybersecurity-specific incident response and crisis management plans that define how a response should be led, managed and coordinated [0]

- challenged to ensure they are effective in a catastrophic ransomware scenario where all IT platforms, cybersecurity tools and usual communication channels are unavailable, and recovery efforts may have to be sustained for weeks or months [0]

###### Business continuity planning and IT disaster recovery planning for a ransomware scenario
<hr width=50%;>

- prioritise business continuity planning and disaster recovery planning that would ensure provision for continuity of critical operations and the ability to recover [0]

i. business continuity planning should be based on rigorous ‘Business Impact Analysis’, and ensure that workarounds are defined for the scenario of total loss of IT for up to several weeks [0]

ii. disaster recovery plans should be based on a prioritised list of applications and systems to recover, should the technology base of the organisation have to be rebuilt or recovered, informed
by an up-to-date asset register and mapping of critical operations to technology [0]

- offline backups (or backups that are verified as inaccessible to attackers with full control of production IT) must be available for all critical systems, data and infrastructure, including core IT infrastructure such as ``Active Directory`` (“AD”) [0]

- with a well-defined and tested restore procedure that includes verification of ability to recover all systems to a common point-in-time [0]

###### Retained incident and crisis support
<hr width=50%;>

- establish contractual retainers with key third parties that may be required to support a crisis response (forensic and technical incident response; crisis response; external legal counsel, and public relations) [0]

## Introduction and background
***

###### Overview of the ransomware cyber attack
<hr width=50%;>

- cyber attack (the ``Incident``) - infiltration of HSE systems using Conti ransomware [0][1]

```python
from IPython.display import IFrame, display
IFrame("./rc/ncsc/HSE_Conti_140521_UPDATE.pdf",width=1000,height=1100)
```

- most serious cyber attacks ever in Ireland, one which affected almost every part of the critical infrastructure of the country’s healthcare system [0]

- reported by HSE management that 80% of the HSE’s environment was encrypted [0]

- hospitals - reverted to using pen and paper [0]

###### Background
<hr width=50%;>

- 2021, the HSE had a combined operating and capital budget of nearly €22bn [0]

- HSE employs approximately 130,000 people, including those directly employed by the HSE as well as those employed by organisations funded by the HSE - is the largest employer in the State [0]

- HSE undertook a Corporate Centre Review in 2020, in order to examine how the HSE’s corporate services are organised to best support operational services [0]

- no recommendations for major changes to IT or cybersecurity in this review (outside of creating a more centralised governance and risk function with the appointment of an equivalent Chief Risk Officer) [0]

- organisational changes took effect approximately 1 month post the Incident [0]

###### HSE IT Infrastructure
<hr width=50%;>

- manages more than 4,500 servers, over 70,000 end user devices and over 1,000 applications with a team of approximately 350 people [0]

###### National Healthcare Network
<hr width=50%;>

- HSE utilises a NHN (National Healthcare Network) for the connectivity and delivery of critical national health services - accessed by the many organisations within the health services [0]

- primarily a ‘flat’ network to make it easier for staff to access the IT applications they require [0]

- design exposes the HSE to the risk of cyber attacks from other organisations connected to the NHN, as well as exposing other organisations to cyber attacks originating from the HSE [0]

- once an attacker has a presence on the network they have ‘freedom to roam’ [0]

- complex IT environment including a significant number of legacy systems [0]

- time of the Incident just under two thirds of the server estate were described as modern (software that is in mainstream support from the vendor) [0]

- remaining estate deemed end of life and operating on extended support or out of support [0]

- had over 30,000 Windows 7 workstations that were deemed end of life by the vendor [0]

- assessed its cybersecurity maturity rating as low (do not have a CISO or a Security Operation Centre established) [0]

## Focus areas - key findings and recommendation
***

- HSE’s Incident Response provider’s investigation determined the ransomware attack originated from a malware infection on a workstation (“patient zero”) on 18 March 2021 [0]

- interacted with a malicious Microsoft Office Excel document that was attached to a phishing email received on 16 March 2021 [0]

< [Table of Contents](#top) | [References](#references) >
<a id="focusareaskeyfindingsandrecommendationAttackerUNC2633"></a>
***
###### ``UNC2633``
***

- patient zero was targeted with phishing emails with the same email subject on four other occasions between 14 December 2020 and 9 February 2021, but the workstation was not successfully infected with malware [0]

- HSE’s Incident Response provider attributed the phishing emails sent to patient zero to an Attacker they refer to as UNC2633 (threat actor that sends phishing emails containing malicious attachments or links to gain access to organisations’ networks) [0]

- additional workstations that were infected with malware (as a result of phishing emails sent by the Attacker UNC2633), with the earliest identified infection occurring on 24 November 2020 (no evidence to indicate these infections contributed to the ransomware incident) [0]

< [Table of Contents](#top) | [References](#references) >
<a id="focusareaskeyfindingsandrecommendationAttackerUNC2727"></a>
***
###### ``UNC2727``
***

- financially motivated group that the HSE’s Incident Response provider has tracked since April 2021 - known to deploy ``Conti`` ransomware and / or exfiltrates victim data in support of their extortion efforts [0]

- after this Attacker gained unauthorised access to the HSE’s IT environment on 18 March 2021, a second Attacker (``UNC2727``) continued to operate in the environment until the execution of ransomware on 14 May 2021 [0]

- compromised 180 systems and at least ``[REDACTED]`` highly privileged accounts (typically required for performing administrative tasks across eight organisations and 19 domains [0]

- identified over 2,800 servers and 3,500 workstations across 15 domains, with evidence of encryption [0]

- likely represents a lower bound on the number of systems encrypted, as some systems were restored from backups or rebuilt prior to endpoint agent deployment, reducing the Incident Response provider’s ability to determine if encryption had occurred [0]

- reference to ‘the Attacker’, will be to the activity attributed to this second Attacker [0]

### Human-operated ransomware attacks
***

- investigating human-operated ransomware attacks it is common to identify evidence of two Attackers:


i. specialises in gaining access to organisations;


ii. specialises in deploying ransomware and extortion [0]

- human-operated ransomware attacks are different to traditional ransomware attacks in that they are ‘hands-on-keyboard’ attacks [0]

- involves human attackers using knowledge of offensive techniques and weaknesses in enterprise IT systems, to methodically compromise organisations’ networks, compromise systems, overcome defence and cause maximum impact [0]

- the ransomware attack that impacted HSE is an example of a human-operated ransomware attack [0]
